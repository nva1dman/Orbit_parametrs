```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import emcee
import corner

# Чтение данных
data = pd.read_csv("4049_60.dat", delim_whitespace=True, header=None, names=["JD", "RV", "Error"])

# Переменные
initial_period = 427  # Исходный период
period_range = np.arange(426, 438, 0.5)  # Диапазон периодов для проверки

# Функция для удаления выбросов
def remove_outliers(x, y, yerr, threshold=3):
    mean_y = np.mean(y)
    std_y = np.std(y)
    mask = np.abs(y - mean_y) < threshold * std_y
    return x[mask], y[mask], yerr[mask]

# Модель эллиптической орбиты
def elliptical_rv_model(phase, gamma, K, omega, e):
    return gamma + K * np.cos(2 * np.pi * phase + omega) / (1 + e * np.cos(2 * np.pi * phase))

# Функция для подбора орбиты с различными параметрами
def fit_orbit(period, threshold=3):
    # Расчет фазы
    phase = np.mod(data['JD'], period) / period
    sorted_indices = np.argsort(phase)
    sorted_phase = phase[sorted_indices]
    sorted_rv = data['RV'][sorted_indices]
    sorted_error = data['Error'][sorted_indices]

    # Удаление выбросов
    sorted_phase, sorted_rv, sorted_error = remove_outliers(sorted_phase, sorted_rv, sorted_error, threshold=threshold)

    # Начальные приближения для параметров
    initial_guess_elliptical = [np.mean(sorted_rv), np.abs((np.max(sorted_rv) - np.min(sorted_rv)) / 2), np.pi, 0.036]

    # Ограничения для параметров
    bounds = ([-np.inf, 0, 0, 0], [np.inf, np.inf, 2 * np.pi, 1])

    # Подгонка модели с учетом начальных условий и ограничений
    params_no_sigma_elliptical, _ = curve_fit(elliptical_rv_model, sorted_phase, sorted_rv, p0=initial_guess_elliptical, bounds=bounds)

    return params_no_sigma_elliptical, sorted_phase, sorted_rv, sorted_error

# Функция для расчета MCMC
def run_mcmc(params, sorted_phase, sorted_rv, sorted_error, nsteps=3000):
    # Логарифмическая функция правдоподобия
    def log_likelihood(theta, phase, rv, rv_err):
        gamma, K, omega, e = theta
        model = elliptical_rv_model(phase, gamma, K, omega, e)
        sigma2 = rv_err**2
        return -0.5 * np.sum((rv - model)**2 / sigma2 + np.log(sigma2))

    # Функция приоритета
    def log_prior(theta):
        gamma, K, omega, e = theta
        if 0 < K < 100 and 0 < omega < 2 * np.pi and 0 < e < 1:
            return 0.0
        return -np.inf

    # Логарифм полной вероятности
    def log_probability(theta, phase, rv, rv_err):
        lp = log_prior(theta)
        if not np.isfinite(lp):
            return -np.inf
        return lp + log_likelihood(theta, phase, rv, rv_err)

    # Начальная позиция и параметры для MCMC
    pos = params + 1e-4 * np.random.randn(32, 4)
    nwalkers, ndim = pos.shape

    # Запуск MCMC
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(sorted_phase, sorted_rv, sorted_error))
    sampler.run_mcmc(pos, nsteps, progress=True)

    return sampler

# Основная функция для тестирования различных периодов
def test_periods_and_fit_mcmc(period_range, threshold=3, nsteps=3000):
    best_period = None
    best_fit_params = None
    best_mcmc_sampler = None
    best_phase, best_rv, best_error = None, None, None

    for period in period_range:
        print(f"Проверка периода: {period:.5f}")
        try:
            params_no_sigma_elliptical, sorted_phase, sorted_rv, sorted_error = fit_orbit(period, threshold)
            
            # Запуск MCMC для текущего периода
            sampler = run_mcmc(params_no_sigma_elliptical, sorted_phase, sorted_rv, sorted_error, nsteps=nsteps)
            
            # Проверим качество подгонки (например, по средней логарифмической вероятности)
            flat_samples = sampler.get_chain(discard=1000, thin=15, flat=True)
            log_prob = sampler.get_log_prob(discard=1000, thin=15, flat=True)
            avg_log_prob = np.mean(log_prob)
            print(f"Средняя логарифмическая вероятность для периода {period:.5f}: {avg_log_prob:.2f}")
            
            # Обновим лучший результат
            if best_fit_params is None or avg_log_prob > np.mean(best_mcmc_sampler.get_log_prob(discard=1000, thin=15, flat=True)):
                best_period = period
                best_fit_params = params_no_sigma_elliptical
                best_mcmc_sampler = sampler
                best_phase, best_rv, best_error = sorted_phase, sorted_rv, sorted_error

        except Exception as e:
            print(f"Ошибка при проверке периода {period:.5f}: {e}")

    # Вернем параметры для лучшего периода
    return best_period, best_fit_params, best_mcmc_sampler, best_phase, best_rv, best_error

# Тестирование периодов
best_period, best_fit_params, best_mcmc_sampler, best_phase, best_rv, best_error = test_periods_and_fit_mcmc(period_range)

# Построение углового графика
samples = best_mcmc_sampler.get_chain(discard=1000, thin=15, flat=True)
labels = ["gamma", "K", "omega", "e"]
fig = corner.corner(samples, labels=labels, truths=best_fit_params)
plt.show()

# Построение лучшей кривой для лучшего периода
gamma_mcmc, K_mcmc, omega_mcmc, e_mcmc = map(lambda v: (v[1], v[2]-v[1], v[1]-v[0]), zip(*np.percentile(samples, [16, 50, 84], axis=0)))
phase_fine = np.linspace(0, 1, 1000)
rv_fitted = elliptical_rv_model(phase_fine, gamma_mcmc[0], K_mcmc[0], omega_mcmc[0], e_mcmc[0])

plt.figure(figsize=(14, 6))
plt.errorbar(best_phase, best_rv, yerr=best_error, fmt='o', ecolor='red', color='black', capsize=7, markersize=6, label='Data')
plt.plot(phase_fine, rv_fitted, 'b-', label=f'Fitted Model (Период: {best_period:.5f} дней)')
plt.xlabel('Orbital Phase')
plt.ylabel('RV (km/s)')
plt.legend()
plt.show()
```

```python
# Получение параметров из результатов MCMC
gamma_mcmc, K_mcmc, omega_mcmc, e_mcmc = map(lambda v: (v[1], v[2]-v[1], v[1]-v[0]), zip(*np.percentile(samples, [16, 50, 84], axis=0)))

# Преобразование омеги в градусы
omega_mcmc_deg = omega_mcmc[0] * (180 / np.pi)
omega_error_deg = (omega_mcmc[1] * (180 / np.pi), omega_mcmc[2] * (180 / np.pi))

# Постоянные
G = 6.67430e-11  # Гравитационная постоянная
day_to_sec = 86400  # Количество секунд в дне
AU_to_m = 1.496e+11  # Астрономическая единица в метрах

# Период в секундах
updated_period_sec = best_period * day_to_sec

# Функция массы
f_m_updated = (np.abs(K_mcmc[0]) * 1e3)**3 * updated_period_sec / (2 * np.pi * G * (1 - e_mcmc[0]**2)**(3/2))
f_m_solar_updated = np.abs(f_m_updated / 1.989e+30)  # Преобразование в солнечные массы

# Ошибка для T0 и f(m)
omega_error_days_updated = omega_error_deg[0] / 360 * best_period
error_f_m = f_m_solar_updated * np.sqrt((3 * K_mcmc[1] / np.abs(K_mcmc[0]))**2 + (3 * omega_error_days_updated / best_period)**2)

# Вычисление T0
phase_at_t0_updated = (np.pi / 2 - omega_mcmc[0]) / (2 * np.pi) % 1
additional_crossing_updated = (3 * np.pi / 2 - omega_mcmc[0]) / (2 * np.pi) % 1

if abs(phase_at_t0_updated - 0.5) < abs(additional_crossing_updated - 0.5):
    phase_at_t0_updated = phase_at_t0_updated
else:
    phase_at_t0_updated = additional_crossing_updated

hjd_at_t0_updated = data['JD'].min() + phase_at_t0_updated * best_period

# Ошибка T0
t0_error_updated = omega_error_days_updated

# Вывод результатов с учетом ошибок MCMC
updated_results_with_mcmc_errors = {
    'Gamma (системная скорость)': f"{gamma_mcmc[0]:.3f} ± {gamma_mcmc[1]:.3f}",
    'K (полуамплитуда)': f"{np.abs(K_mcmc[0]):.3f} ± {K_mcmc[1]:.3f}",
    'Omega (градусы)': f"{omega_mcmc_deg:.3f} ± {omega_error_deg[0]:.3f}",
    'e (эксцентриситет)': f"{e_mcmc[0]:.3f} ± {e_mcmc[1]:.3f}",
    'T0 (HJD)': f"{hjd_at_t0_updated:.3f} ± {t0_error_updated:.3f} дней",
    'Функция массы (f(m))': f"{f_m_solar_updated:.6f} ± {error_f_m:.6f} солнечных масс"
}

# Печать результатов
for key, value in updated_results_with_mcmc_errors.items():
    print(f"{key}: {value}")
```
